```r
library(ggplot2)
library(ggrepel)
library(EnhancedVolcano)
```

```r
res=read.csv("/home/data1/Asthma/EnhancedVolcano/data.csv",sep = "\t")
```

```r
res
```

```r
 p1<-EnhancedVolcano(res,
    lab = res$Assay,
    x = 'restimate',
    y = 'p.value',
    selectLab = res[res$p.value<=0.01,]$Assay,
    xlab="Effect size (CT vs CC)",
    title = '',
    subtitle = "",
    pCutoff = 0.05,
    FCcutoff = 0,
    pointSize =  -log(res$p.value)*0.8,
labCol = 'black',
    labFace = 'bold',
    colAlpha = 0.9,
    colGradient =  c('#FFBE7A',"#9ac9db"),
    caption = "",
    drawConnectors = TRUE,
    widthConnectors = 0.5,
    labSize = 4.0)
```

```r
pdf("/home/data1/Asthma/EnhancedVolcano/EnhancedVolcano_plot.pdf", width = 6, height = 8)

p1+ggplot2::coord_cartesian(ylim=c(0, 3),xlim=c(-2, 2))+ 
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank(),legend.position="none",axis.text = element_text(size =1),axis.title = element_text(size =15))
dev.off()
```
